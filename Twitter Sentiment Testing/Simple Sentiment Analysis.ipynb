```python
tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
        
```

```python
tweets = pd.DataFrame(tweets_data)
# tweets
```

```python
tweet_data = pd.DataFrame(tweets['text'])
tweet_data.sort_values('text')
```

```python
# Split

#twitter tokenizer


tweet_data['text_lc'] = tweet_data['text'].str.lower()
tweet_data['text_token'] = tweet_data['text_lc'].apply(word_tokenize)
tweet_data['text_token']
```

```python
#Calculate and print the total number of personal pronouns and possessive pronouns used by female and male authors
tweet_data['tokens_pos'] = tweet_data['text_token'].apply(lambda x: [word for word in x if word in pos_list])
tweet_data['pos_count'] = tweet_data['tokens_pos'].apply(len)
tweet_data
```

```python
# Create a new column that contains the proportion of total words that are personal pronouns for each row
tweet_data['pos_count_prop'] = tweet_data['pos_count']/tweet_data['text_token'].apply(len)
tweet_data
```

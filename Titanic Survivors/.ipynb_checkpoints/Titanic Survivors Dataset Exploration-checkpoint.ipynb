## Titanic Survivors Dataset Exploration

This notebook is an exploration of Kaggle's Titanic Survior's dataset. The challenge is to predict the survival rate of passengers.

This is <em><style font-size=24>currenly a work-in-progress</style></em>!

#### Variables
Variable	Definition	Key <br>
survival	Survival	0 = No, 1 = Yes  <br>
pclass	Ticket class	1 = 1st, 2 = 2nd, 3 = 3rd  <br>
sex	Sex  <br>
Age	Age in years  <br>
sibsp	# of siblings / spouses aboard the Titanic  <br>
parch	# of parents / children aboard the Titanic  <br>
ticket	Ticket number  <br>
fare	Passenger fare  <br>
cabin	Cabin number  <br>
embarked	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton

```python
#importing libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%pylab inline
```

```python
train_df = pd.read_csv('dataset/train.csv')
train_df
```

```python
#graphing
input_values = list(train_df['Survived'][:100])
input_values
index = list(train_df['PassengerId'][:100])

plt.scatter(index, input_values, marker=".")

# Set chart title and label axis
plt.title("Titanic Survivors", fontsize=20)
plt.xlabel("Index", fontsize=14)
plt.ylabel("Survived", fontsize=14)

sns.set() #will switch to seaborn defaults
plt.show()
```
